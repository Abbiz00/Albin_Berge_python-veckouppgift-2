```python
print("Program som kommer räkna ut om du får åka med på Balder eller ej")
while True:
    try:
        length= float(input("Skriv in din längd i cm"))
        break
    except ValueError:
        print("Du måste skriva in en siffra")

if length <= 121:
    print("Tyvärr du är för kort för att åka med")
else:
    print("Du är välkommen att åka med! ")
```

3 Sportresultat
Tottenham spelar mot Liverpool i Champions League. Skriv ett program som frågar användaren hur många mål respektive lag gjorde, och talar om vilket lag som vann.
Exempel på output:

Matchen är över, nu ska vi räkna ut resultatet!
Hur många mål gjorde Tottenham? 2
Hur många mål gjorde Liverpool? 1
Tottenham vann!


```python
print("Program som räknar ut vilket lag som har vunnit")
while True:
    try:
        goal_tottenham = int(input("Hur många mål gjorde Tottenham i matchen?"))
        break
    except ValueError:
        print("Du måste ange en siffra!")

while True:
    try:
        goal_Liverpool = int(input("Hur många mål gjorde Liverpool i matchen?"))
        break
    except ValueError:
        print("Du måste ange en siffra")

if goal_tottenham > goal_Liverpool:
    win_tottenham = True
else: win_tottenham = False

if goal_tottenham < goal_Liverpool:
    win_liverpool = True
else: win_liverpool = False



print("Matchen är över, nu skall vi räkna ut resultatet!")
print(f"Tottenham gjorde {goal_tottenham} mål")
print(f"Liverpool gjorde {goal_Liverpool} mål")
if win_tottenham:
    print(f"Tottenham vann!")
if win_liverpool:
    print(f"Liverpool vann!")

```
